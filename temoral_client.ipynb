## using workflow_service

See also https://docs.temporal.io/search-attribute 

### count workflows

````python
from temporalio.api.workflowservice.v1 import CountWorkflowExecutionsRequest
req = CountWorkflowExecutionsRequest(namespace='default', query='WorkflowId = "pay-invoice-702"')
resp = await service.count_workflow_executions(req)
print(resp)
````

### list workflows

````python
from temporalio.api.workflowservice.v1 import ListWorkflowExecutionsRequest
req = ListWorkflowExecutionsRequest(namespace='default', query='WorkflowId="pay-invoice-702"')
resp = await service.list_workflow_executions(req)
print(resp)
````
